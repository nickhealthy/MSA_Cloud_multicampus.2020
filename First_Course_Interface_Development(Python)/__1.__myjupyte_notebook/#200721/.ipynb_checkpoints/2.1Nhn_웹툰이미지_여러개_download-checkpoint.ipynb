#### 특정 웹툰 페이지 모든 image를 다운로드 하기
* soup.select('img[src$=.jpg]')
* img 폴더를 생성하고 img 폴더 하위에 파일 저장
<br>

```python
import requests
from bs4 import BeautifulSoup
import os

main_url= 'https://comic.naver.com/webtoon/detail.nhn?titleId=626906&no=602&amp;weekday=tue'
res = requests.get(main_url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')

img_urls = []
for img_url in soup.select("img[src$='.jpg']"):
    #print(img_url)
    img_urls.append(img_url['src'])
    
# 디렉토리 생성
if not os.path.isdir('img'):
    os.mkdir('img')

print(len(img_urls)) 
for img_url in img_urls:
    #print(img_url)
    req_header = {
        'referer':main_url
    }
    
    res2 = requests.get(img_url, headers=req_header)
    img_data = res2.content
    file_name = os.path.basename(img_url)
    with open('img/'+file_name,'wb') as file:
        print('Writing to {} ({} bytes)'.format(file_name,len(img_data)))    
        file.write(img_data)    
```

### 특정 웹툰
<br>
<br>
<br>

```python
def write_image(title, main_url):
    print(title, main_url)
    import requests
    from bs4 import BeautifulSoup
    import os

    # main_url= 'https://comic.naver.com/webtoon/detail.nhn?titleId=626906&no=602&amp;weekday=tue'
    res = requests.get(main_url) # 통신연결을 해
    html = res.text # HTML 소스를 가져와
    soup = BeautifulSoup(html, 'html.parser') # WHITE SPACE 지우고 >> HTML 코드만 PARSER

    img_urls = []            # 리스트 안에 들어갈 딕셔너리 >> 이중 리스트화 변수 설정
    # soup.select 자체가 list 화 됌 >> 'img 태그' 안에 'src 속성' 값중 뒤에 >> .jpg 글자만 가져옴
    for img_url in soup.select("img[src$='.jpg']"): 
        #print(img_url)
        img_urls.append(img_url['src'])              # img_urls 리스트 안에 img_url['src 속성']에 있는 글자 >> .jpg 만 리스트화

    # 디렉토리 생성
    dir_name = 'img/' + title                          
    if not os.path.isdir(dir_name):            # if 없으면 img 폴더 안에 title 변수에 적은 이름 폴더 생성
        os.mkdir(dir_name)

    print(len(img_urls))             #img_urls 리스트 개수
    for img_url in img_urls:          
        #print(img_url)
        req_header = {                # 해당 페이지의 헤더 주소 값 설정
            'referer':main_url
        }
        res2 = requests.get(img_url, headers=req_header)   # requests.get 통신 연결 >> header 값만
        img_data = res2.content                                 # img_data 변수에 binary 값으로 저장
        file_name = os.path.basename(img_url)                   
        with open(dir_name+'/'+file_name,'wb') as file:
            print('Writing to {} ({} bytes)'.format(file_name,len(img_data)))    
            file.write(img_data)    
    
    
```

```python
write_image('마음의소리', 'https://comic.naver.com/webtoon/detail.nhn?titleId=20853&no=1236')
```

```python
# 내가 한거

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# http 통신 밑 HTML 요청하기
main_url = 'https://comic.naver.com/index.nhn'
res = requests.get(main_url)
html = res.text

# parsing 작업
soup = BeautifulSoup(html, 'html.parser')

# 
# print('a 태그 개수', len(soup.select('.genreRecomInfo2 h6 a'))) 

webtoon_list = []
webtoon_dict = {}


for div_tag in soup.select('.genreRecomInfo2'):
    for div_tag_h6_a in div_tag.select('h6 a'):
        title = div_tag_h6_a.text.strip()
        link = urljoin(main_url, div_tag_h6_a['href'])
        webtoon_dict['title'] = title
        webtoon_dict['link'] = link
        
    for div_tag_p_a in div_tag.select('p a'):
        link =  urljoin(main_url, div_tag_p_a['href'])
        webtoon_dict['detail_link'] = link
    webtoon_list.append(webtoon_dict)
    webtoon_dict = {}
    
webtoon_list
```
