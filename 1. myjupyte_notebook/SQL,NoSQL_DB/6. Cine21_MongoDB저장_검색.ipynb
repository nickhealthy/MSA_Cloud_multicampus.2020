```python
import os
import pymongo
import json

# 1. connection 생성 - MongoDB랑 연결
conn = pymongo.MongoClient(host='localhost', port=27017)
print(conn)

# 2. database 생성
movie = conn.movie_db # conn.movie_db << database name
print(movie)
#print(movie.name) # database name 확인

# 3. collection 생성
actors = movie.actor_col
print(actors)
#print(actors.name) # collection name 확인

# 4. json file open
# with open('data/actors.json', 'r', encoding='utf8') as file:
#     json_data = json.load(file)
#     actors.insert_many(json_data)

# document(row) count
print(actors.estimated_document_count()) # estimated_document_count() << document 의 개수
```

# 1. sort

```python
docs = actors.find({})
for idx, actor in enumerate(docs, 1):
    #print(actor) # actor를 뿌리면 dict type이라 모든 값이 나오기 때문에 밑에처럼 원하는 값만 뿌린다.
    print(idx, actor['배우이름'], actor['흥행지수'])
```

```python
# 기본적으로 ascending
docs = actors.find({}).sort('배우이름')
# decending
docs = actors.find({}).sort('배우이름', pymongo.DESCENDING)
# sort by multiple field
docs = actors.find({}).sort([('배우이름', pymongo.ASCENDING), ('랭킹', pymongo.DESCENDING)])
for idx, actor in enumerate(docs, 1):
    #print(actor)
    print(idx, actor['배우이름'], actor['랭킹'], actor['흥행지수'])
```

# 2. exists
* 필드값이 존재하는 경우에 한해서 검색

```python
# '특기' 필드가 존재하는 경우와 존재하지 않는 경우 count
print(actors.find({'특기':{'$exists' : True}}).count())
print(actors.find({'특기':{'$exists' : False}}).count())
```

```python
# 생년월일이 있는 배우의 이름과 생년월일 (exists 사용하지 않고)
# find({},{'배우이름':1, '생년월일':1,'_id':0}}) << {조건},{보고싶은 필드}

#docs = actors.find({},{'배우이름':1, '생년월일':1,'_id':0})
docs = actors.find({'생년월일':{'$exists':True}}) # 생년월일이 있는 애들만 뿌려줌

for actor in docs:
    #print(actor) # 이건 오류 안남
    print(actor['배우이름'], actor['생년월일']) # 특정 컬럼에 값이 없는 것들이 있을 수도 있으니 exists 를 사용하는것이 안전
```

```python
# '다른 이름'이 '조원준' 인 배우이름과 다른이름 랭킹을 출력하세요
for actor in actors.find({'다른 이름':{'$exists':True},'다른 이름':'조원준'}):
    print(actor['배우이름'], actor['다른 이름'], actor['랭킹'])
```

# 3. 필드의 범위로 검색
* $gt,  $gte
* $lt,  $lte
* filter 조건 사용
* fields 따로 변수로 지정

```python
# 흥행지수 10000 <= 흥행지수 <= 30000 배우를 검색
# count()
# 배우이름:1, 흥행지수:1, 출연영화:1, _id:0

filter = {'흥행지수':{'$gte':10000, '$lte':30000}}
fields = {'배우이름':1, '흥행지수':1, '출연영화':1, '_id':0}

print(actors.find(filter).count()) # 19건
for actor in actors.find(filter, fields): # 필터 = 조건, 필드 = 보고 싶은 필드
    print(actor)
```

# 4.or, nor (not or)

```python
# or
# 흥행지수가 >= 15000 이거나 랭킹 <= 20 인 배우이름과 출연영화 리스트 출력하기

match = {'$or' : [{'흥행지수':{'$gte':15000}}, {'랭킹':{'$lte':20}}]}
field = {'배우이름' : 1, '흥행지수':1, '랭킹':1, '출연영화' : 1, '_id' : 0}
for doc in actors.find(match, field):
    print(doc)
```

```python
# $nor = (not or)
# 흥행지수 >= 15000, 성별 = 남

match = {'$nor' :[{'흥행지수':{'$lte':15000}}, {'성별':'여'}]}
field = {'배우이름' : 1, '흥행지수':1, '랭킹':1, '출연영화' : 1, '_id' : 0}
# print(actors.find(match).count()) # 4명
for actor in actors.find(match, field):
    print(actor)
```

# 5. $sortByCount
* aggregate() 함수 사용

```python
# 소속사별 count
match = [{'$sortByCount': '$소속사'}] # aggregate 쓸땐 [] 로 먼저 묶어줘야함
match = [{'$sortByCount': '$소속사'}, {'$match':{'count':{'$gte':4}}}]

for actor in actors.aggregate(match):
    print(actor)
```

# 6. list 검색
* 리스트 안에 들어 있는 내용을 검색하기 위해서는 순서와 값이 정확히 일치해야 함
* 순서에 상관없이 값으로만 찾으려면 &all 를 사용한다

```python
# 출연영화 리스트에 '반도' 를 포함하는 경우
match = {'출연영화':'반도'}

# 출연영화 리스트에 '1987'또는 '반도' 를 포함하는 경우
match = {'$or': [{'출연영화':'1987'}, {'출연영화':'반도'}]}

# 출연영화 리스트에 '남산의 부장들' 과 '공작' 을 포함하는 경우
match = {'출연영화':['남산의 부장들', '공작']} # 조건에 만족하지 않는다. >> 못찾음(순서와 개수가 맞아야함)
# $all 사용
match = {'출연영화':{'$all':['남산의 부장들', '공작']}} # 조건에 만족한다.

# 출연영화의 첫번째 요소가 '결백' 인 경우
match = {'출연영화.0':'결백'} # 조건1).0 인덱스를 지정 >> 첫번째 값을 찾는 조건을 지정해주고 조건2) '결백' 를 준다.
# 출연영화의 두번째 요소가 '밀양' 인 경우
match = {'출연영화.1':'밀양'}

# 출연영화의 갯수가 6편인 경우
match = {'출연영화':{'$size':6}}

field = {'배우이름':1, '출연영화':1, '_id':0}
for actor in actors.find(match, field):
    print(actor)
```

# 7. limit()

```python
# 직업이 가수이면서 흥행지수가 높은 상위 10명의 배우는?
for actor in actors.find({'직업':'가수'}).sort('흥행지수', pymongo.DESCENDING).limit(10):
    print(actor)
```

```python
# '국가부도의 날'에 출연한 배우를 흥행지수가 높은 순으로 10명을 출력하세요
# '_id' : 0

for actor in actors.find({'출연영화':'국가부도의 날'}, {'_id':0}).sort('흥행지수', pymongo.DESCENDING).limit(10):
    print(actor)
```
